```python
import polars as pl
from skyulf.profiling.analyzer import EDAAnalyzer
from skyulf.profiling.visualizer import EDAVisualizer

# 1. Load Data
df = pl.read_csv("../customer_churn_dataset.csv")
df = df.drop("customer_id")
# 2. Run Analysis
analyzer = EDAAnalyzer(df)
profile = analyzer.analyze(
    target_col="churn",  # Changed from "target" to "churn"
    task_type="Classification",
)

# 3. Visualize Results (The Easy Way)
viz = EDAVisualizer(profile, df)

# Print the dashboard
viz.summary()

# Show the plots
viz.plot()
```
